```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Longitudinal
x = np.arange(0, 0.3, .01)

A = 25
B = 85
P = 1.5

F = (B*x)/(1 + np.abs(A*x)**P)

plt.plot(x, F)
plt.show()

targetValue = .2

print(max(F))
print((B*targetValue)/(1 + np.abs(A*targetValue)**P))
print(F[-1])

# Lateral
x = np.arange(0, 15, .1)

A = 1.5
B = 4
P = 1.2

F = (B*x)/(1 + np.abs(A*x)**P)

plt.plot(x, F)
plt.show()

print(max(F))
print(F[-1])
```

```python
x = np.arange(0, .3, .001)

shape = 100
mupeak = 1.7
muslip = 1.5
alphapeak = 0.1
alphaslip = 1.25
linearBreak = alphapeak*0.75
slope = (-shape*(linearBreak - alphapeak)**2 + mupeak)/linearBreak

y1 = slope*x
ygradient = -(2*x - 2*alphapeak)
y2 = -shape*(x - alphapeak)**2 + mupeak
y3 = 1/(x) + muslip

plt.plot(x, y1, label = "linear")
plt.plot(x, y2, label = "Parabola")
plt.plot(x, y3, label = "Asymptote")
plt.legend()
plt.ylim(0, 2)
plt.show()
```
